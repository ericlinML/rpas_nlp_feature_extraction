```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
import statsmodels.api as sm
```

```python
df = pd.read_csv('rpasNLPamar.csv')
```

```python
df.columns
```

```python
protocolList = ['gloveCosSim Avg Protocol',
       'gloveWordPairwiseCosSimSBERT Avg Protocol',
       'gloveWordPairwiseCosSim6b300d Avg Protocol',
       'gloveWordPairwiseCosSim840b300d Avg Protocol',
       'word2vecWordPairwiseCosSim Avg Protocol', 'robertaCosSim Avg Protocol',
       'robertaWordPairwiseCosSim Avg Protocol',]
```

```python
for item in protocolList:
    sns.histplot(df[item])
    plt.show()
```

```python
for index, item in zip(df.columns, df.isna().sum()):
    if item != 0:
        print(index, item)
```

```python
for item in protocolList:
    print(item)
    sm.qqplot(df[item], line='45', fit=True)
    plt.show()
```

```python
tempList = protocolList.copy()
```

```python
for item in tempList:
    print(item)
    tempList.remove(item)
    for item2 in tempList:
        sns.jointplot(data=df, x=item, y=item2, kind='reg')
        print(f"{item} {item2}")
        plt.show()
    print('')
```

# Pearson's R

```python
for item in tempList:
    print(item)
    tempList.remove(item)
    for item2 in tempList:
        print(f"{item} {item2}")
        print(stats.pearsonr(df[item], df[item2]))
        print('')
    print('')
    print('')
```

# Spearman's rho

```python
tempList = protocolList.copy()
```

```python
for item in tempList:
    print(item)
    tempList.remove(item)
    for item2 in tempList:
        print(f"{item} {item2}")
        print(stats.spearmanr(df[item], df[item2]))
        print('')
    print('')
    print('')
```

# Kendall's Tau

```python
tempList = protocolList.copy()
```

```python
for item in tempList:
    print(item)
    tempList.remove(item)
    for item2 in tempList:
        print(f"{item} {item2}")
        print(stats.kendalltau(df[item], df[item2], method='exact'))
        print('')
    print('')
    print('')
```
